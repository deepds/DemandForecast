```python
import numpy as np
import pandas as pd
import featuretools as ft
```

```python
from pandas.tseries.offsets import Day, Week
```

```python
week = pd.read_parquet('/main/data/MVideo_2/week_ft_5.parquet').reset_index()
```

```python
week['ID'] = week.STORE_ID.astype('str') + ' ' + week.PRODUCT_ID.astype('str')
```

```python
week['KEY'] = week.DATE.astype('str') + ' ' + week.STORE_ID.astype('str') + ' ' + week.PRODUCT_ID.astype('str')
```

```python
week['DUMMY'] = 0
```

```python
week.shape
```

```python
cutoff = week[['index', 'DATE']]
```

```python
ds = week[['KEY', 'DATE']]
```

```python
ind = week[['ID']].drop_duplicates()
```

```python
store = week[['STORE_ID']].drop_duplicates()
```

```python
prod = week[['PRODUCT_ID']].drop_duplicates()
```

```python
cat = week[['SUBCATEGORY_ID']].drop_duplicates()
```

```python
trans = pd.read_parquet('/main/data/MVideo_2/transactions_ft_5.parquet').reset_index()
```

```python
trans['ID'] = trans.STORE_ID.astype('str') + ' ' + trans.PRODUCT_ID.astype('str')
```

```python
trans.info()
```

```python
trans['time_index'] = trans.DATE
```

```python
hol = pd.read_parquet('/main/data/MVideo_2/hol_ft_5.parquet')
```

```python
hol.info()
```

```python
hol['DUMMY'] = 0
```

```python
hol['time_index'] = hol.DATE
```

```python
hol.shape
```

```python
cal = hol[['DUMMY']].drop_duplicates()
```

```python
es = ft.EntitySet(id = 'data')
```

```python
variable_types_1 = {
    'REGULAR_SALES_PRICE_INCL_TAX': ft.variable_types.Categorical,
    'PRODUCT_ID': ft.variable_types.Categorical,
    'MAINCATEGORY_ID': ft.variable_types.Categorical,
    'SUBCATEGORY_ID': ft.variable_types.Categorical,
    'STORE_ID': ft.variable_types.Categorical,
    'REGION': ft.variable_types.Categorical,
    'INTERNET': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W0': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W1': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W2': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W3': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W4': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W5': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W6': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W7': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_W8': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_P1': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_ABS_DIFF_MEAN_8': ft.variable_types.Categorical,
    'PROMOTIONS_AVAILABLE_DIFF_P1': ft.variable_types.Categorical,
}
```

```python
es = es.entity_from_dataframe(entity_id = 'week', 
                              dataframe = week,
                              variable_types = variable_types_1,
                              index = 'index',
                              time_index = 'DATE')
```

```python
es['week']
```

```python
es = es.entity_from_dataframe(entity_id = 'sales', 
                              dataframe = ind,
                              index = 'ID',
                             )
```

```python
es = es.entity_from_dataframe(entity_id = 'store', 
                              dataframe = store,
                              index = 'STORE_ID',
                             )
```

```python
es = es.entity_from_dataframe(entity_id = 'product', 
                              dataframe = prod,
                              index = 'PRODUCT_ID',
                             )
```

```python
es = es.entity_from_dataframe(entity_id = 'category', 
                              dataframe = cat,
                              index = 'SUBCATEGORY_ID',
                             )
```

```python
variable_types_2 = {
    'STORE_ID': ft.variable_types.Categorical,
    'PRODUCT_ID': ft.variable_types.Categorical,
    'SUBCATEGORY_ID': ft.variable_types.Categorical
}
```

```python
es = es.entity_from_dataframe(entity_id = 'trans', 
                              dataframe = trans,
                              variable_types = variable_types_2,
                              index = 'index',
                              time_index = 'time_index')
```

```python
es = es.entity_from_dataframe(entity_id = 'holidays', 
                              dataframe = hol,
                              index = 'DATE',
                              time_index = 'time_index')
```

```python
es = es.entity_from_dataframe(entity_id = 'general', 
                              dataframe = cal,
                              index = 'DUMMY'
                             )
```

```python
es = es.add_relationship(ft.Relationship(es['sales']['ID'], es['week']['ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['sales']['ID'], es['trans']['ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['store']['STORE_ID'], es['week']['STORE_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['store']['STORE_ID'], es['trans']['STORE_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['product']['PRODUCT_ID'], es['week']['PRODUCT_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['product']['PRODUCT_ID'], es['trans']['PRODUCT_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['category']['SUBCATEGORY_ID'], es['week']['SUBCATEGORY_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['category']['SUBCATEGORY_ID'], es['trans']['SUBCATEGORY_ID']))
```

```python
es = es.add_relationship(ft.Relationship(es['general']['DUMMY'], es['week']['DUMMY']))
```

```python
es = es.add_relationship(ft.Relationship(es['general']['DUMMY'], es['holidays']['DUMMY']))
```

```python
es
```

```python
es.add_last_time_indexes()
```

```python
# 167 hours is 6 days and 23 hours
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="week",
                                      cutoff_time = cutoff,
                                      training_window = ft.Timedelta(167, 'h'),
                                      agg_primitives = ['sum', 'std', 'max', 'min', 'percent_true'],
                                      trans_primitives = ['week', 'month'],
                                      chunk_size = 0.05,
                                      n_jobs = 21,
                                      verbose = True)
```

```python
feature_defs
```

```python
res = feature_matrix.copy()
```

```python
res.drop(columns = ['ID', 'DUMMY'], inplace = True)
```

```python
res1 = pd.merge(ds, res, how = 'inner', on = ['KEY'])
```

```python
res1.drop(columns = ['KEY'], inplace = True)
```

```python
res1
```

```python
res1.fillna(0, inplace = True)
```
